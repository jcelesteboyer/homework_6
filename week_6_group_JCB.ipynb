#### Import jeopardy.csv and create a dataframe from the data. 

```python
import pandas as pd
import numpy as np
file_to_load = "JEOPARDY_CSV.csv"
jeopardy_csv = pd.read_csv(file_to_load)
```

```python
jeopardy_df = pd.DataFrame(jeopardy_csv)
jeopardy_df.head()
```

How many rows are there in the dataframe? What does a row represent in this dataset? What is the "grain"?

```python
# There are 216930 rows in this dataframe. 
# A row represents a question that was asked in the show 'Jeopardy'
# and includes other valuable data like the answer, value, and show number. 
# The 'grain' of the dataset is one question/answer combo per episode and air date.
# And reports a round, catergory, and value. 
print(len(jeopardy_df))
```

Clean the data. How do you need to "clean"it? What is "dirty" about this data? 

```python
# I am checking to see if there is any missing values.
# Length changes from 216930 to 216928 rows which tells me that there might be two rows with missing data. 
jeopardy_df.dropna()
```

```python
null_data = jeopardy_df[jeopardy_df.isnull().any(axis=1)]
null_data

# We can see that two of the answers in the dataframe is NaN 
# but it looks like the real answer was 'Null' and maybe python mistook that for missing information. 
```

```python
jeopardy_df = jeopardy_df.fillna('Null')
#I filled in the the NaN values with the correct answer of 'Null'.
# Now to make sure that this has worked: 
jeopardy_df.dropna()
```

```python
# The next step in cleaning up the data is to drop duplicates where full rows are duplicated. 
# Because we can have have duplicated answers for pretty much every single column, 
#however, it would be unlikely that the same question was asked twice in an episode.
jeopardy_df.groupby(' Air Date')[' Question'].agg(pd.Series.drop_duplicates)
print(len(jeopardy_df))
#There were no duplicated rows in this dataframe. 
```

How many unique show were there each year?

```python
jeopardy_df[' Air Date'] = pd.to_datetime(jeopardy_df[' Air Date'])
jeopardy_df['Year'] = jeopardy_df[' Air Date'].dt.year

unique_shows = pd.DataFrame(jeopardy_df.groupby('Year')['Show Number'].nunique())

unique_shows.head()
```

How many times was each category used? 

```python
category_counts_df = pd.DataFrame(jeopardy_df[' Category'].value_counts())
category_counts_df.head()
```

Which month (over the years) has had the most shows?

```python
jeopardy_df['Month'] = jeopardy_df[' Air Date'].dt.month
mon_most_shows_df = pd.DataFrame(jeopardy_df.groupby('Month')['Show Number'].nunique())
mon_most_shows_df.columns = ['Shows Per Month']
mon_most_shows_df.sort_values('Shows Per Month',ascending=False).head(1)
#The month with the most number of shows over the years is November. 
```

Are there any answers that appeared multiple times? Find the questions for the top three most common answers?

```python
answer_counts_df = pd.DataFrame(jeopardy_df[' Answer'].value_counts())
answer_counts_df.head(3)
# This is the top three most common answers. 
```

Create a visualization that shows the top 5 categories (those reference the most often). Be sure to label appropriately. 

```python
import matplotlib.pyplot as plt
# Creating a df for the top five catergories
jeopardy_top_5_cat_df = pd.DataFrame(jeopardy_df[' Category'].value_counts())
jeopardy_top_5_cat_df.head()
```

```python
# Sorting the catergories first before cutting off the rows not needed. 
jeopardy_top_5_cat_df2 = pd.DataFrame(jeopardy_top_5_cat_df.sort_values(by=[' Category'],ascending=False))
# I need to now cut off the rest of the rows that I don't need for this plot. 
top_5_df = pd.DataFrame(jeopardy_top_5_cat_df2.drop(jeopardy_top_5_cat_df.index[5:]))

top_5_df.plot(kind='bar',xlabel='Category',ylabel='No. of References',title='Top 5 Jeopardy Categories')

plt.xticks(rotation = 45)
plt.show()
```
